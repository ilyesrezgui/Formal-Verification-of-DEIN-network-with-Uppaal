```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Define the data as multi-line strings
data1 = """
0.0,0.0
3.275501900012079,0.0
3.275501900012079,13.0
604800.0,13.0
"""

data2 = """
0.0,0.0
1.355368342615772,0.0
3.0012730767757505,13.0
3.0012730767757505,124.0
604800.0,124.0
"""

data3 = """
0.0,0.0
1.0354549697833315,0.0
1.0354549697833315,33.0
604800.0,33.0
"""

data4 = """
0.0,0.0
0.35624402650799203,0.0
1.519082336307704,71.0
1.519082336307704,182.0
604800.0,182.0
"""

data5 = """
0.0,0.0
0.433544435426363,0.0
5.515220549613285,101.0
6.273111567608256,212.0
7.00541386733475,245.0
9.318633880071435,356.0
9.318633880071435,467.0
604800.0,467.0
"""

# Convert multi-line strings to pandas DataFrames
df1 = pd.read_csv(StringIO(data1), header=None, names=['Time', 'EnergyLost'])
df2 = pd.read_csv(StringIO(data2), header=None, names=['Time', 'EnergyLost'])
df3 = pd.read_csv(StringIO(data3), header=None, names=['Time', 'EnergyLost'])
df4 = pd.read_csv(StringIO(data4), header=None, names=['Time', 'EnergyLost'])
df5 = pd.read_csv(StringIO(data5), header=None, names=['Time', 'EnergyLost'])

# Calculate the sum of energy lost for each dataset
sum_energy_lost = {
    'EWD=2': df1['EnergyLost'].sum(),
    'EWD=10': df2['EnergyLost'].sum(),
    'EWD=30': df3['EnergyLost'].sum(),
    'EWD=60': df4['EnergyLost'].sum(),
    'EWD=90': df5['EnergyLost'].sum()
}

# Convert dictionary to lists for plotting
labels = list(sum_energy_lost.keys())
sizes = list(sum_energy_lost.values())

# Initialize a figure
plt.figure(figsize=(10, 6))

# Plot the bar chart
plt.bar(labels, sizes, color='skyblue')

# Add title and labels
plt.title('Sum of Energy Lost per Wireless Device (WD)')
plt.xlabel('Wireless Device (WD)')
plt.ylabel('Sum of Energy Lost')

# Show the plot
plt.show()
```
